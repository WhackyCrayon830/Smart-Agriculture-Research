```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://agmarknet.gov.in/viewmarketprofileinputpublic")

time.sleep(10)  # React hydration

# ---- CLICK STATE DROPDOWN ----
driver.execute_script("""
let labels = document.querySelectorAll('label');
for (let lbl of labels) {
    if (lbl.innerText.includes('State')) {
        let row = lbl.closest('div.flex');
        let box = row.querySelector('div.cursor-pointer');
        box.click();
        break;
    }
}
""")

print("State dropdown clicked")

time.sleep(1)

# ---- SELECT FIRST STATE OPTION (SCOPED) ----
state = driver.execute_script("""
let searchBox = document.querySelector("input[placeholder='Search...']");
if (!searchBox) return 'SEARCH_BOX_NOT_FOUND';

let dropdown = searchBox.parentElement;
let options = dropdown.querySelectorAll("div.cursor-pointer");

if (options.length === 0) return 'NO_STATE_OPTIONS';

options[0].click();
return options[0].innerText;
""")

print("Selected state:", state)

input("Press Enter to close browser...")
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://agmarknet.gov.in/viewmarketprofileinputpublic")

time.sleep(10)  # React hydration

# ---------- STATE ----------
driver.execute_script("""
for (let lbl of document.querySelectorAll('label')) {
    if (lbl.innerText.includes('State')) {
        lbl.closest('div.flex').querySelector('div.cursor-pointer').click();
        break;
    }
}
""")

time.sleep(1)

state = driver.execute_script("""
let s = document.querySelector("input[placeholder='Search...']");
let opts = s.parentElement.querySelectorAll("div.cursor-pointer");
opts[0].click();
return opts[0].innerText;
""")

print("Selected state:", state)

# ---------- DISTRICT ----------
time.sleep(3)

driver.execute_script("""
for (let lbl of document.querySelectorAll('label')) {
    if (lbl.innerText.includes('District')) {
        lbl.closest('div.flex').querySelector('div.cursor-pointer').click();
        break;
    }
}
""")

time.sleep(1)

district = driver.execute_script("""
let s = document.querySelector("input[placeholder='Search...']");
let opts = s.parentElement.querySelectorAll("div.cursor-pointer");
opts[0].click();
return opts[0].innerText;
""")

print("Selected district:", district)

# ---------- MARKET ----------
time.sleep(3)

driver.execute_script("""
for (let lbl of document.querySelectorAll('label')) {
    if (lbl.innerText.includes('Market')) {
        lbl.closest('div.flex').querySelector('div.cursor-pointer').click();
        break;
    }
}
""")

time.sleep(1)

markets = driver.execute_script("""
let s = document.querySelector("input[placeholder='Search...']");
let dropdown = s.parentElement;
let opts = dropdown.querySelectorAll("div.cursor-pointer");

let names = [];
for (let o of opts) {
    let t = o.innerText.trim();
    if (t.length > 0) names.push(t);
}
return names;
""")

print(f"Markets found ({len(markets)}):")
for m in markets:
    print(" -", m)

# ---------- SAVE ----------
with open("markets.txt", "w", encoding="utf-8") as f:
    for m in markets:
        f.write(m + "\n")

with open("markets.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["State", "District", "Market"])
    for m in markets:
        writer.writerow([state, district, m])

print("\nSaved markets.txt and markets.csv")

input("Press Enter to close browser...")
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://agmarknet.gov.in/viewmarketprofileinputpublic")

time.sleep(10)  # React hydration

rows = []

# ---------- HELPERS ----------
def open_dropdown(label_text):
    driver.execute_script(f"""
    for (let lbl of document.querySelectorAll('label')) {{
        if (lbl.innerText.includes('{label_text}')) {{
            lbl.closest('div.flex').querySelector('div.cursor-pointer').click();
            break;
        }}
    }}
    """)

def read_dropdown_options():
    return driver.execute_script("""
    let s = document.querySelector("input[placeholder='Search...']");
    if (!s) return [];
    let opts = s.parentElement.querySelectorAll("div.cursor-pointer");
    return Array.from(opts).map(o => o.innerText.trim()).filter(t => t);
    """)

def read_selected_value(label_text):
    return driver.execute_script(f"""
    for (let lbl of document.querySelectorAll('label')) {{
        if (lbl.innerText.includes('{label_text}')) {{
            return lbl.closest('div.flex').querySelector('div.cursor-pointer').innerText.trim();
        }}
    }}
    return '';
    """)

def select_option_by_index(idx):
    driver.execute_script(f"""
    let s = document.querySelector("input[placeholder='Search...']");
    if (!s) return;
    let opts = s.parentElement.querySelectorAll("div.cursor-pointer");
    if (opts.length > {idx}) opts[{idx}].click();
    """)

# ---------- STATES ----------
open_dropdown("State")
time.sleep(1)
states = read_dropdown_options()
open_dropdown("State")  # close

current_state = read_selected_value("State")
print(f"Initial state: {current_state}")

# ---------- PROCESS INITIAL STATE ----------
time.sleep(3)

open_dropdown("District")
time.sleep(1)
districts = read_dropdown_options()
open_dropdown("District")

current_district = read_selected_value("District")
print(f"  Initial district: {current_district}")

time.sleep(3)

open_dropdown("Market")
time.sleep(1)
markets = read_dropdown_options()
open_dropdown("Market")

if not markets:
    rows.append([current_state, current_district, ""])
else:
    for m in markets:
        rows.append([current_state, current_district, m])

# ---------- LOOP REMAINING STATES ----------
for si in range(1, len(states)):
    open_dropdown("State")
    time.sleep(1)
    select_option_by_index(si)
    time.sleep(3)

    state = states[si]
    print(f"\nSTATE [{si+1}/{len(states)}]: {state}")

    open_dropdown("District")
    time.sleep(1)
    districts = read_dropdown_options()
    open_dropdown("District")

    # ---------- PROCESS FIRST DISTRICT ----------
    open_dropdown("District")
    time.sleep(1)
    select_option_by_index(0)
    time.sleep(3)

    district = districts[0]

    open_dropdown("Market")
    time.sleep(1)
    markets = read_dropdown_options()
    open_dropdown("Market")

    if not markets:
        rows.append([state, district, ""])
    else:
        for m in markets:
            rows.append([state, district, m])

    # ---------- LOOP REMAINING DISTRICTS ----------
    for di in range(1, len(districts)):
        open_dropdown("District")
        time.sleep(1)
        select_option_by_index(di)
        time.sleep(3)

        district = districts[di]
        print(f"  DISTRICT [{di+1}/{len(districts)}]: {district}")

        open_dropdown("Market")
        time.sleep(1)
        markets = read_dropdown_options()
        open_dropdown("Market")

        if not markets:
            rows.append([state, district, ""])
        else:
            for m in markets:
                rows.append([state, district, m])

# ---------- SAVE ----------
with open("india_mandis.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["state", "district", "market"])
    writer.writerows(rows)

print("\n✅ DONE — NO STATES OR DISTRICTS MISSED")
print(f"Total rows: {len(rows)}")

input("Press Enter to close browser...")
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv
import os

OUTPUT_FILE = "india_mandis.csv"

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://agmarknet.gov.in/viewmarketprofileinputpublic")

time.sleep(10)  # React hydration

# ---------- HELPERS ----------
def open_dropdown(label_text):
    driver.execute_script(f"""
    for (let lbl of document.querySelectorAll('label')) {{
        if (lbl.innerText.includes('{label_text}')) {{
            lbl.closest('div.flex').querySelector('div.cursor-pointer').click();
            break;
        }}
    }}
    """)

def read_dropdown_options():
    return driver.execute_script("""
    let s = document.querySelector("input[placeholder='Search...']");
    if (!s) return [];
    let opts = s.parentElement.querySelectorAll("div.cursor-pointer");
    return Array.from(opts).map(o => o.innerText.trim()).filter(t => t);
    """)

def read_selected_value(label_text):
    return driver.execute_script(f"""
    for (let lbl of document.querySelectorAll('label')) {{
        if (lbl.innerText.includes('{label_text}')) {{
            return lbl.closest('div.flex').querySelector('div.cursor-pointer').innerText.trim();
        }}
    }}
    return '';
    """)

def select_option_by_index(idx):
    driver.execute_script(f"""
    let s = document.querySelector("input[placeholder='Search...']");
    if (!s) return;
    let opts = s.parentElement.querySelectorAll("div.cursor-pointer");
    if (opts.length > {idx}) opts[{idx}].click();
    """)

# ---------- CSV SETUP ----------
file_exists = os.path.exists(OUTPUT_FILE)

csv_file = open(OUTPUT_FILE, "a", newline="", encoding="utf-8")
writer = csv.writer(csv_file)

if not file_exists:
    writer.writerow(["state", "district", "market"])
    csv_file.flush()

# ---------- READ STATES ----------
open_dropdown("State")
time.sleep(1)
states = read_dropdown_options()
open_dropdown("State")

current_state = read_selected_value("State")
print(f"Initial state: {current_state}")

# ---------- PROCESS INITIAL STATE ----------
time.sleep(3)

open_dropdown("District")
time.sleep(1)
districts = read_dropdown_options()
open_dropdown("District")

current_district = read_selected_value("District")
print(f"  Initial district: {current_district}")

time.sleep(3)

open_dropdown("Market")
time.sleep(1)
markets = read_dropdown_options()
open_dropdown("Market")

if not markets:
    writer.writerow([current_state, current_district, ""])
else:
    for m in markets:
        writer.writerow([current_state, current_district, m])

csv_file.flush()  # 🔥 WRITE NOW
print(f"  ✔ Saved {current_state} / {current_district}")

# ---------- LOOP REMAINING STATES ----------
for si in range(1, len(states)):
    open_dropdown("State")
    time.sleep(1)
    select_option_by_index(si)
    time.sleep(3)

    state = states[si]
    print(f"\nSTATE [{si+1}/{len(states)}]: {state}")

    open_dropdown("District")
    time.sleep(1)
    districts = read_dropdown_options()
    open_dropdown("District")

    # ---------- FIRST DISTRICT ----------
    open_dropdown("District")
    time.sleep(1)
    select_option_by_index(0)
    time.sleep(3)

    district = districts[0]

    open_dropdown("Market")
    time.sleep(1)
    markets = read_dropdown_options()
    open_dropdown("Market")

    if not markets:
        writer.writerow([state, district, ""])
    else:
        for m in markets:
            writer.writerow([state, district, m])

    csv_file.flush()  # 🔥 WRITE NOW
    print(f"  ✔ Saved {state} / {district}")

    # ---------- REMAINING DISTRICTS ----------
    for di in range(1, len(districts)):
        open_dropdown("District")
        time.sleep(1)
        select_option_by_index(di)
        time.sleep(3)

        district = districts[di]
        print(f"  DISTRICT [{di+1}/{len(districts)}]: {district}")

        open_dropdown("Market")
        time.sleep(1)
        markets = read_dropdown_options()
        open_dropdown("Market")

        if not markets:
            writer.writerow([state, district, ""])
        else:
            for m in markets:
                writer.writerow([state, district, m])

        csv_file.flush()  # 🔥 WRITE NOW
        print(f"    ✔ Saved {state} / {district}")

# ---------- CLEANUP ----------
csv_file.close()
driver.quit()

print("\n✅ DONE — CSV UPDATED LIVE")
print(f"File: {OUTPUT_FILE}")
```
